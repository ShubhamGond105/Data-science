```python
import numpy as np
import pandas as pd
```

```python
employees = pd.DataFrame({
    "emp_id": [101, 102, 103, 104, 105],
    "name": ["Amit", "Priya", "Rahul", "Sneha", "Vikram"],
    "department": ["HR", "IT", "Finance", "IT", "Marketing"]
})

# DataFrame 2: Employee Salaries
salaries = pd.DataFrame({
    "emp_id": [101, 102, 103, 106],
    "salary": [45000, 60000, 55000, 70000],
    "bonus": [5000, 7000, 6000, 8000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='emp_id',how='inner')
```

```python
pd.merge(employees,salaries,on='emp_id',how='outer')
```
